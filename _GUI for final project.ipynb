```python
import tkinter as tk
from tkinter import *
import joblib

# Create the main window
root = tk.Tk()
root.title("Maternal Health Risk Assessment")

# declaring IntVar variables for input fields
age_entry = tk.IntVar()
systolic_bp_entry = tk.IntVar()
diastolic_bp_entry = tk.IntVar()
bs_entry = tk.IntVar()
body_temp_entry = tk.IntVar()
heart_rate_entry = tk.IntVar()
entries = [age_entry, systolic_bp_entry, diastolic_bp_entry, bs_entry, body_temp_entry, heart_rate_entry]

# Load the pre-trained Random Forest model and scaler
clf = joblib.load('finalModel.joblib')
sc = joblib.load('scaler.joblib')

# Define the target mapping dictionary
target_mapping = {0: 'low risk', 1: 'mid risk', 2: 'high risk'}

# Function to handle form submission
def submit_form():
    age = age_entry.get()
    systolic_bp = systolic_bp_entry.get()
    diastolic_bp = diastolic_bp_entry.get()
    bs = bs_entry.get()
    body_temp = body_temp_entry.get()
    heart_rate = heart_rate_entry.get()

    # Prepare the input data as a feature vector
    inp = [age, systolic_bp, diastolic_bp, bs, body_temp, heart_rate]

    # Scale the input data using the loaded scaler
    scaled_inp = sc.transform([inp])

    # Use the loaded model to make predictions
    rf_predict = clf.predict(scaled_inp)

    # Map the prediction to a risk level
    predicted_class = target_mapping[rf_predict[0]]

    # Display the predicted class in your Tkinter interface
    myLabel.config(text=f"Predicted Class: {predicted_class}")

# Create and pack labels and Entry widgets for health data inputs
data_labels = ["Age:", "Systolic BP:", "Diastolic BP:", "Blood Sugar (BS):", "Body Temperature:", "Heart Rate:"]
data_entries = []

for label_text in data_labels:
    label = tk.Label(root, text=label_text)
    label.pack()

for _, ent in zip(data_labels, entries):
    entry = tk.Entry(root, textvariable=ent)
    entry.pack()

# Create a Submit button
submit_button = tk.Button(root, text="Submit", command=submit_form)
submit_button.pack()

# Label for displaying predictions
myLabel = Label(root, text="")
myLabel.pack()

# Start the Tkinter main loop
root.mainloop()
```
